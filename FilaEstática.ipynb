Fila estática



```python
class Fila:
    """
    Fila estática circular
    """
    def __init__(self, capacity):
        
        self.capacity = capacity
        self.front = -1
        self.rear = -1 
        self.size = 0
        self.data = [None] * capacity
    def isEmpty(self):
        if self.size == 0:
            return True
        else:
            return False
    #return self.size == 0  Faz a mesma coisa que o if
    
    def isFull(self):
        if self.size >= self.capacity:
            return True
        else: 
            return False
        
    def enqueue(self, valor):
        if self.isEmpty(): 
            self.front = 0 
        
        if self.isFull():
            print("Erro (overflow)")
            return
        #atualiza quem é o ultimo 
        if self.rear == self.capacity -1:
            self.rear = 0
        else:
            self.rear += 1
        
        self.data[self.rear] = valor
        self.size += 1
        
    def dequeue(self):
        if self.isEmpty():
            print("Erro (underflow)")
            return
        valorExcluido = self.data[self.front]
        if self.size == 1:
            #atualiza quem é o último
            self.rear = -1
            self.front = -1
       
        
        elif self.front == self.capacity -1:
            self.front = 0 
        else:
        #atualiza quem é o primeiro
            self.front += 1
        self.size += -1 #atualiza o tamanho da fila
        return valorExcluido
    
    def __str__(self):
        if self.isEmpty():
            return "[]"
        else:
            strFila = "["
        
            filaAux = Fila( capacity = self.capacity )
            while not self.isEmpty():
                valor = self.dequeue()
                strFila += " " + str(valor)
                
                filaAux.enqueue(valor)
            strFila += " ] "
            while not filaAux.isEmpty():
                valor = filaAux.dequeue()
                self.enqueue(valor)
        return strFila
# testa a fila
fila = Fila(capacity = 6)
print(fila)
fila.enqueue(3)
fila.enqueue(5)
fila.enqueue(7)
fila.enqueue(8)
fila.enqueue(9)
fila.enqueue(11)
print(fila)
valor = fila.dequeue()
print(fila)
valor = fila.dequeue()
print(fila)
valor = fila.enqueue(55)
print(fila)
valor = fila.enqueue(99)
print(fila)
valor = fila.dequeue()
print(fila)
valor = fila.dequeue()
print(fila)
```
