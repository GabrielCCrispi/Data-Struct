```python
class Carro:
    """
    Classe que simula um carro
    """
    
    def __init__(self, cor = "Azul", 
                       velocidade = 0.0,
                       marca = None):
        """
        Método construtor
        """
        
        self.cor = cor
        self.velocidade = velocidade
        
        if marca is None:
            print('\n\nVocê não passou a marca.')
            print('Criarei automático.')
            
            if self.cor == "Azul":
                self.marca = "Fiat"
            else:
                self.marca = "Chevrolet"
        else:
            self.marca = marca
        
    def acelerar(self):
        self.velocidade += 10
        
    def imprimir(self):
        print('\nCor: ', self.cor)
        print('Veloc.: ', self.velocidade)
        print('Marca: ', self.marca)
        
# instanciar a classe Carro
carro1 = Carro()
carro1.imprimir()

help(carro1)

carro2 = Carro("Vermelho")
carro2.imprimir()

carro3 = Carro( velocidade = 50 )
carro3.imprimir()

carro4 = Carro( velocidade = 100, cor="Marrom" )
carro4.imprimir()

carro5 = Carro(marca = "Renault")
carro5.imprimir()
```

**Ex01.** Crie uma classe chamada Calculadora. 
- Essa classe deve receber dois números e passar para dois atributos: num1 e num2.
- Crie quatro métodos que usem esses atributos: soma, substração, divisão e multiplicação. 

    - Esses métodos não devem imprimir o resultado, mas apenas retornar

```python
class Calculadora:
    
    def __init__(self, num1, num2):
        
        self.num1 = num1
        self.num2 = num2
        
    def soma(self):
        res = self.num1 + self.num2
        return res

    def subtracao(self):
        res = self.num1 - self.num2
        return res    

    def multiplicacao(self):
        res = self.num1 * self.num2
        return res 
    
    def divisao(self):
        res = self.num1 / self.num2
        return res 
    
num1 = input("Digite o primeiro valor: ")
num2 = input("Digite o segundo valor: ")

num1 = int(num1)
num2 = int(num2)

# instancia a calculadora
calc1 = Calculadora(num1, num2)
resSoma = calc1.soma()
print("Soma: ", resSoma)

resSub = calc1.subtracao()
print("Sub: ", resSub)

resDiv = calc1.divisao()
print("Div: ", resDiv)

resMul = calc1.multiplicacao()
print("Mult: ", resMul)
```
